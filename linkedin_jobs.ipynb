## 5- Remove the word "followers" from (linkedin_followers) and convert it into a numerical type
#### Description : To make more use of the column convert it to integer after removing the unnecessary text

```python
df_final['linkedin_followers'].value_counts()
```

```python
#remove 'followers'
df_final['linkedin_followers'] = df_final['linkedin_followers'].str.replace('followers','')
#remove the commas ','
df_final['linkedin_followers'] = df_final['linkedin_followers'].str.replace(',','').str.strip() 
#fill the nulls with zeros to be able to convert them
#df_final['linkedin_followers'] = df_final['linkedin_followers'].fillna('0')
#replace all non-numeric values with zeros
df_final['linkedin_followers'] = df_final['linkedin_followers'].replace('[^0-9]+', np.nan, regex=True) 
#convert to float (because it accepts null values)
df_final['linkedin_followers'] = df_final['linkedin_followers'].astype(float)
df_final.info()
```
